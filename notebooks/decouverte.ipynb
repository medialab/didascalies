## 1. quels groupes parlent dans le dataset ?

```python
from vega import VegaLite
```

```python
VegaLite({
    "mark": "bar",
    "transform": [{
        "filter": "datum.parlementaire_groupe_acronyme !== null"
    },{
        "filter": "datum.intervention.length > 30"
    }],
    "encoding": {
        "x": {
            "type": "nominal",
            "field": "parlementaire_groupe_acronyme"
        },
        "color": {
            "type": "nominal",
            "field": "parlementaire_groupe_acronyme"
        },
        "y": {
            "field": "*",
            "aggregate": "count",
            "type": "quantitative"
        }
    }
}, df)
```

```python
VegaLite({
    "mark": "point",
    "transform": [{
        "filter": "datum.sexe !== null"
    }],
    "encoding": {
        "x": {
            "type": "nominal",
            "field": "sexe"
        },
        "color": {
            "type": "nominal",
            "field": "sexe"
        },
        "y": {
            "type": "nominal",
            "field": "fonction"
        },
        "size": {
            "field": "*",
            "aggregate": "count",
            "type": "quantitative"
        }
    }
}, df)
```
