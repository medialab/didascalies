```python
import re
import csv
from itertools import islice
```

```python
INPUT = '../data/interventions.csv'
```

```python
with open(INPUT, 'r') as f:
    INTERVENTIONS = [line for line in csv.DictReader(f)]
```

```python
# Helpers
TAGS_RE = re.compile(r'<[^>]+>')
GOOD_RE = re.compile(r'\bbien\b|\bvrai\b|\boui\b|\bmerci\b|\braison\b|\bfavorable\b')

def strip_tags(html):
    return re.sub(TAGS_RE, '', html)

def is_interruption(intervention):
    intervention = strip_tags(intervention.lower())
    
    if len(intervention) > 30:
        return False
    
    if 'séance' in intervention:
        return False
    
    if GOOD_RE.search(intervention):
        return False
    
    return True
```

```python
INTERRUPTIONS = {'H': {'H': 0, 'F': 0}, 'F': {'H': 0, 'F': 0}}
H = 0
F = 0
for i, intervention in islice(enumerate(INTERVENTIONS), None):
    if intervention['sexe'] == 'NULL':
        continue
        
    if intervention['sexe'] == 'H':
        H += 1
    else:
        F += 1
        
    if is_interruption(intervention['intervention']):
        continue
        
    content = strip_tags(intervention['intervention'])

    next_intervention = INTERVENTIONS[i + 1]

    if next_intervention['sexe'] == 'NULL':
        continue

    if not is_interruption(next_intervention['intervention']):
        continue
        
    if intervention['sexe'] == 'F' and next_intervention['sexe'] == 'H':
        print(intervention['intervention'], intervention['parlementaire'])
        print(next_intervention['intervention'])
        print()
        
    INTERRUPTIONS[intervention['sexe']][next_intervention['sexe']] += 1
```

```python
print('H', H)
print('F', F)
R = F / H
print('R', R)
print()
IH = INTERRUPTIONS['H']['H'] + INTERRUPTIONS['H']['F']
IF = INTERRUPTIONS['F']['H'] + INTERRUPTIONS['F']['F']
print('IH', IH)
print('IF', IF)
print()
print(IH / H)
print(IF / F)
print()
print(INTERRUPTIONS['H']['H'] / IH)
print(INTERRUPTIONS['H']['F'] / IH)
print()
print(INTERRUPTIONS['F']['H'] / IF)
print(INTERRUPTIONS['F']['F'] / IF)
print()
print((IH + IF) / (F + H))
print(IH / H)
print(IF / F)
```
