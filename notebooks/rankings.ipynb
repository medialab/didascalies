```python
%matplotlib inline

import matplotlib.pyplot as plt

def top_reaction(data, col, keyword, speaker):
    if col == "didascalie":
        diz={"parlementaire":"nom", "nom":"parlementaire"}
        d = data[(data[diz[speaker]].isnull())]
        d[col] = d[col].str.lower()
        dat = d[d[col].str.contains(keyword)==True]
        print dat.shape
        co = dat[speaker].value_counts()
    else:
        diz={"parlementaire":"nom", "nom":"parlementaire"}
        d = data[(data[diz[speaker]].isnull())]
        d[col] = d[col].str.lower()
        dat = d[d[col].str.contains(keyword)==True]
        print dat.shape
        co = dat[speaker].value_counts()
    return co
```

```python
top_reaction(df,"didascalie","applaud","nom").plot(kind='barh',figsize=(10,10))
```
